```python
fewshot_mapping = {
    'leaderboard_bbh': 3,
    'leaderboard_gpqa': 0,
    'leaderboard_mmlu_pro': 5,
    'leaderboard_musr': 0,
    'leaderboard_ifeval': 0,
    'leaderboard_math_hard': 4,
    'gsm8k': 5,
    'truthfulqa': 0,
    'winogrande': 5,
    'arc_challenge': 25,
    'hellaswag': 10,
    'mmlu':5,
}
```

```python
print("Qwen")

model_name = "Qwen/Qwen3-8B-Base"

for task, num_fewshot in fewshot_mapping.items():
    print("--"*200)
    print("**********Doing TASK NAME: ", task)
    command = f"accelerate launch -m lm_eval --model hf " \
                  f"--model_args pretrained={model_name},trust_remote_code=True,dtype=bfloat16 " \
                  f"--tasks {task} " \
                  f"--num_fewshot {num_fewshot} " \
                  f"--batch_size auto:1"

    print(command)
    # Execute the command
    !{command}

# truthful - 77.86 vs 76.07
# arc - 73.12 vs 73.21
# winograde - 84.53 vs 85.24
```
